# Technologies for Advanced Programming: WashYourCar!



<hr>

L'idea del progetto nasce dall'esigenza di trovare, nell'arco di una settimana, il giorno più adatto da dedicare al lavaggio della propria auto, evitando giorni piovosi che possono susseguire.

## Tecnologie utilizzate

Per la realizzazione della pipeline sono stati utilizzate diverse tecnologie.

La prima tecnologia utilizzata per la fase di data ingestion è Logstash, tramite essa con l'ausilio in input del plug-in logstash-input-http_poller ed in output del plug-in logstash-integration-kafka, vengono acquisiti dati dalla sorgente ad intervalli imposti in fase di configurazione ed inseriti opportunamente all'interno di una coda Kafka.

La seconda tecnologia utilizzata è Apache Kafka, una piattaforma di messaggistica open-source di natura publish-subscribe per la gestione dello streaming di dati in tempo reale e lo storage temporaneo di essi. Le sue prorprietà di scalabilità ed efficienza la rendono una tecnologia ideale per la costruzione di una pipeline, per questi motivi essa verrà utilizzata come destinazione da parte di Logstash che inserirà dati al suo interno in un topic ben definito in fase di configurazione. I dati inseriti nella coda verranno successivamente consumati da Apache Spark che gode di un'ottima integrazione con Kafka, per la fase di elaborazione della nostra pipeline.

La terza tecnologia utilizzata è Apache Spark, un sistema di elaborazione distribuito utilizzato in genere con grandi carichi di lavoro. Grazie alle tecniche di ottimizzazione utilizzate esso permette elevate prestazioni che risultano indispensabili per la fase di elaborazione dati di una pipeline realtime.

La quarta tecnologia utilizzata è Elasticsearch, una tecnologia per la ricerca e l'analisi distribuita di dati, attraverso delle tecniche di indicizzazione, Elasticsearch permette l'archiviazione di documenti JSON successivamente recuperabili attraverso una referenza da esso aggiunta. Elasticsearch offre insieme a Kibana una soluzione ottima per lo storage di documenti JSON e la loro successiva visualizzazione attraverso pannelli di controllo interattivi. 

La quinta tecnologia utilizzata è Kibana, una piattaforma web per la presentazione visiva dei dati raccolti attraverso la nostra pipeline, lavora a stretto contatto con Elasticsearch ed insieme a Logstash forma il cosiddetto Stack ELK. 

Nel nostro caso specifico, i dati verranno immessi da Logstash su Kafka tramite le API fornite da https://openweathermap.org/api. I dati riguarderanno le previsioni meteo di una posizione data da latitudine e longitudine. Attraverso Spark, filtreremo i dati della coda creata da Kafka in modo tale da accedere alle informazioni che servono per il nostro caso d'uso (Temperatura, probabilità di precipitazioni, umidità e percentuale di cielo coperto). Attraverso ElasticSearch e Kibana possiamo, infine, visualizzare i nostri indici e previsioni.

## Let's Start!

Una volta scaricata la cartella bisognerà:

- Eseguire Docker
- Entrare nella cartella del progetto tramite terminale
- Eseguire il comando "docker compose up"
- Aprire il browser e connettersi all'indirizzo "localhost:5601"
- Caricare il file di dashboard "Dashboard.ndjson"
